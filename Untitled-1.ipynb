```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.metrics import mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dense, Flatten, Concatenate
import kagglehub
import random
from sklearn.compose import ColumnTransformer
import kerastuner as kt

```

Cargamos el dataset

```python
# Cargar dataset
path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")
df_esperanza = pd.read_csv(path+'/Life Expectancy Data.csv')
```

```python
# Preprocesar datos
df_esperanza['Country'] = df_esperanza['Country'].astype('category')
```

```python
# Manejar valores NaN
df_esperanza = df_esperanza.dropna()
```

```python
X = df_esperanza.drop(columns=['Life expectancy '])
y = df_esperanza['Life expectancy ']
```

```python
# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
# Preprocesar características categóricas y escalar características numéricas
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object', 'category']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ], remainder='passthrough')

X_train_scaled = preprocessor.fit_transform(X_train)
X_test_scaled = preprocessor.transform(X_test)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))
```

```python
# Establecer semilla para reproducibilidad
np.random.seed(5)
random.seed(5)  
tf.random.set_seed(5)

# Definir función de construcción del modelo para Keras Tuner
def build_model(hp):
    input_country = Input(shape=(1,))
    input_features = Input(shape=(X_train_scaled.shape[1] - 1,))
    
    embedding = Embedding(
    input_dim=len(df_esperanza['Country'].unique()) + 1,  # +1 para estar seguro
    output_dim=hp.Int('embedding_output_dim', min_value=5, max_value=500, step=5)
)(input_country)
    flatten = Flatten()(embedding)
    
    concat = Concatenate()([flatten, input_features])
    
    x = concat
    for i in range(hp.Int('num_layers', 1, 3)):
        x = Dense(units=hp.Int('units_' + str(i), min_value=32, max_value=512, step=32),
                  activation='relu')(x)
    output = Dense(1)(x)
    model = Model(inputs=[input_country, input_features], outputs=output)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Configurar Keras Tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=20,
    factor=3,
    directory='my_dir',
    project_name='life_expectancy'
)

# Buscar los mejores hiperparámetros
tuner.search([X_train['Country'].values, X_train_scaled[:, 1:]], y_train_scaled, epochs=50, validation_split=0.2, verbose=1)

# Obtener el mejor modelo
best_model = tuner.get_best_models(num_models=1)[0]


# Obtener el mejor modelo
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
# Entrenar modelo
model.fit(X_train_scaled, y_train_scaled, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Predecir y evaluar
y_pred_scaled = model.predict(X_test_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
```
